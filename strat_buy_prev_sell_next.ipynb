```python
import numpy
import pandas as pd
import yfinance as yf
```

```python
df = yf.download("ADBE", start="2021-07-12", end="2021-08-12", interval="1d")
```

```python
df['ClosePrev'] = df['Close'].shift(1)
```

```python
num_stocks = 10
df['Prof'] = df['Open']-df['ClosePrev']
df['actual'] = num_stocks*df['Prof']
df['isHighMoreThanClose'] = df['High'] > df['ClosePrev']
df['best'] = num_stocks*(df['High'] - df['ClosePrev'])
df['worked'] = df['isHighMoreThanClose'] | (df['Prof'] > 0)
df
```
