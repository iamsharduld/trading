```python
import yfinance as yf
```

```python
data = yf.download("SPY", start="2021-07-04", end="2021-08-30", interval="1d")
```

```python
x = data['Close'].shift(1)
data['change%'] = 100*((data['Close'] - x)/x)
data['changeHigh'] = 100*((data['High'] - x)/x)
data['changeLow'] = 100*((data['Low'] - x)/x)
data['portfolio'] = 1000

```
